# Data Analysis Notebook

This notebook examines the empirical data gathered from the UCLA Business School on hiring, promotion, and attrition activity of faculty.

## Percentage of Female Faculty in the Department
This first chart shows that there is an improvement in the gender balance within the department over time. We see the gradual rise in female faculty over 12 years. The other trends we see are this rather periodic behavior of gradual upturns follows by downturns, as if a few year of positive female hiring is followed by a few years of no-female hiring. 


**NOTE: PLOT THE PROPORTION OF FEMALE HIRES ON RIGHT Y-AXIS.**

```python
plt.plot(mgmt_data['yr'], mgmt_data['act_fpct'], label='Management Data')
plt.title('Proportion of Female Faculty over the Entire Department')
plt.xlabel('Year')
plt.ylabel('Proportion Women Faculty')
plt.legend()
plt.grid()
plt.show()
```

## Percentage of Female Faculty versus Male Faculty

This chart shows the comparison between female and male faculty percentage in the department. While there is an obvious shrinkage in the gap gender gap, the change seems to be very gradual.

```python
plt.plot(mgmt_data['yr'], mgmt_data['act_fpct'], label='Percent Female Faculty')
plt.plot(mgmt_data['yr'], 1 - mgmt_data['act_fpct'], label='Percent Male Faculty ')
plt.title('Percentage of Male and Female Faculty over Total Department ')
plt.xlabel('Year')
plt.ylabel('Percent Faculty')
plt.legend()
plt.grid()
plt.show()
```

## Cumulative sum of hires



```python
men_hires = np.cumsum(mgmt_data['act_mhire'])
female_hires = np.cumsum(mgmt_data['act_fhire'])

plt.plot(mgmt_data['yr'], female_hires, label='Data - Female Asst')
plt.plot(mgmt_data['yr'], men_hires, label='Data - Male Asst')
plt.title('Men versus Women Hiring Asst ')
plt.xlabel('Year')
plt.ylabel('Number of Hires')
plt.legend()
plt.grid()
plt.show()
```

## Female faculty proportion versus hiring proportion

```python
hiring_prop = mgmt_data['act_fhire']/(mgmt_data['act_fhire'] + mgmt_data['act_mhire'])
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Year')
ax1.set_ylabel('Proportion Female Faculty')
ax1.plot(mgmt_data['yr'], mgmt_data['act_fpct'], label='Female prop', color='k')
ax2 = ax1.twinx()
ax2.set_ylabel('Proportion of Female Hires out of Total Hires')
ax2.plot(mgmt_data['yr'], hiring_prop, label='Hiring prop', color='r')
plt.legend()
plt.grid()
plt.show()
```
