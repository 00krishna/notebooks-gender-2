```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nlower = 64
Nupper = 84
```

```python

Nt = np.linspace(60, 90, 200)
```

```python
p_hire = 0.05
```

```python
y = p_hire + (p_hire/(1+np.exp(Nt - Nlower))) - (p_hire/(1+np.exp(Nupper - Nt)))
```

```python
plt.plot(Nt, y)
plt.title("Probability of Position Creation between Upper/Lower Dept. Bounds")
plt.xlabel('Department Size')
plt.ylabel('Probability of Position Creation')
plt.savefig('test.png')
```

```python
y2 = (p_hire/(1+np.exp(Nt - Nlower)))
```
